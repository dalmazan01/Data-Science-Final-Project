```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# === Load your data ===
# Replace this with your actual file path
file_path = '/workspaces/Data-Science-Final-Project/NBA/data/all_team_cps_oddh.csv'
df = pd.read_csv(file_path)

# === Clean and convert relevant columns ===
df['Closing_Spread'] = pd.to_numeric(df['Closing_Spread'], errors='coerce')
#df['Win_PCT'] = df['Win_PCT'].str.replace('%', '').astype(float)
df['Average_MOV'] = pd.to_numeric(df['Average_MOV'], errors='coerce')
#df['Cover_PCT'] = df['Cover_PCT'].str.replace('%', '').astype(float)

# Drop any rows with missing data
df = df.dropna(subset=['Closing_Spread', 'Win_PCT', 'Average_MOV', 'Cover_PCT'])

# === Create correlation matrix ===
corr_matrix = df[['Closing_Spread', 'Win_PCT', 'Average_MOV', 'Cover_PCT']].corr()

# === Plot heatmap ===
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Between Spread, Win %, MOV, and Cover %')
plt.tight_layout()
plt.show()
```
