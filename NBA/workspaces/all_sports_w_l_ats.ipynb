```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Placeholder file paths for actual CSVs
nba_path = '/workspaces/Data-Science-Final-Project/NBA/data/w_l_nba_ats.csv'
nfl_path = '/workspaces/Data-Science-Final-Project/NBA/data/w_l_nfl_ats.csv'
mlb_path = '/workspaces/Data-Science-Final-Project/NBA/data/w_l_mlb_ats.csv'

# Load the data
nba_df = pd.read_csv(nba_path)
nfl_df = pd.read_csv(nfl_path)
mlb_df = pd.read_csv(mlb_path)

# Extract relevant values from each dataset (assuming one row each for now)
nba_team = nba_df.iloc[0]
nfl_team = nfl_df.iloc[0]
mlb_team = mlb_df.iloc[0]

# Construct a unified DataFrame for analysis
data = {
    'League': ['NBA', 'NFL', 'MLB'],
    'Team': [nba_team['Team'], nfl_team['Team'], mlb_team['Team']],
    'Win_PCT': [float(nba_team['Win_PCT']), float(nfl_team['Win_PCT']), float(mlb_team['Win_PCT'])],
    'MOV': [float(nba_team['MOV']), float(nfl_team['MOV']), float(mlb_team['MOV'])],
    'ATS_RunLine_PlusMinus': [float(nba_team['ATS_+/-']), float(nfl_team['ATS_+/-']), float(mlb_team['Run_Line_+/-'])]
}
df = pd.DataFrame(data)

# Plot 1: Win % Comparison
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='League', y='Win_PCT', palette='Blues_d')
plt.title('Team Win % by League')
plt.ylabel('Win Percentage')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Margin of Victory (MOV) Comparison
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='League', y='MOV', palette='Greens_d')
plt.title('Margin of Victory (MOV) by League')
plt.ylabel('Average Point/Run Margin')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 3: ATS / Run Line +/- Comparison
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='League', y='ATS_RunLine_PlusMinus', palette='coolwarm')
plt.title('ATS / Run Line +/- by League')
plt.ylabel('Average Cover Margin')
plt.grid(True)
plt.tight_layout()
plt.show()

df
```
